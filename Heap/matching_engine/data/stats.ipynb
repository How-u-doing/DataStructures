```python
import duckdb

df = duckdb.query("""
    select *
    from read_csv('orders.csv', names=['id', 'party', 'price', 'quantity', 'timestamp', 'side'])
""").df()
df
```

```python
xf = duckdb.query("""
    select party, sum(price * quantity * if(side='BUY', 1, -1)) as position
    from read_csv('orders.csv', names=['id', 'party', 'price', 'quantity', 'timestamp', 'side'])
    group by party
""").df()
xf
```
